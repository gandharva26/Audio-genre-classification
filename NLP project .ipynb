```python
import pandas as pd

df = pd.read_csv('train (1).csv',encoding="UTF-8") #text articles
print (df)


# In[3]:


import pandas as pd

df1 = pd.read_excel (r'Book1.xlsx')#stop words
print (df1)


# In[4]:


#for classifing the text we need to convert it into numbers but before that we need to delete stop words:
c=df1['stopwords']
x=[]
for i in range(0,len(c)):
    x.append(c[i])


# In[5]:


c=df['headline']
p=df['label']
c1=[]
p1=[]
for i in range(0,len(c)):
    c1.append(c[i]) #article
    p1.append(p[i])# topics


# In[6]:


c1


# In[7]:


def stop(text):
        stop_words = set(x) #remove kannada stop words
        import re
        
        word_tokens = text.split()
        filtered_sentence = [w for w in word_tokens if not w in stop_words] 
        filtered_sentence = [] 
        
        bad_chars = ['\u200b','\u200d']

        for k in range(33,127):
            bad_chars.append(chr(k)) #contains all ascii chars from 33 to 127
    
        #print(word_tokens)
        for j in range(0,len(word_tokens)): 
            if j not in stop_words: 
                
                test_string = word_tokens[j]
                test_string = ''.join(i for i in test_string if not i in bad_chars) 
                jk=str(test_string)
                if(len(jk)!=0):
                    filtered_sentence.append(jk) 
        #print(word_tokens) 
        p=(filtered_sentence) 
        # print(p)
        return p
    
gf=[]
for i in range(0,len(c1)):
    gf.append(stop(c1[i]))


# In[8]:


f=sum(gf,[])#combine list

output = []
for x in f:
    if x not in output:
        output.append(x)

# Create the pandas DataFrame 
af = pd.DataFrame( columns = ['topic']+output) 


#a=['a','b','c','d']
#b=[['a','b'],['b'],['d','c']]
a=output
b=gf
aa=[]
for i in range(0,len(b)):
        xx=[]
        for m in range(0,len(a)):
            xx.append(0)
        for j in range(0,len(b[i])):
            for k in range(0,len(a)):
                if(b[i][j]==a[k]):
                    yy=a.index(a[k])
                    xx[yy]=1
        aa.append(xx)   
    
import csv
with open('nlppro.csv', 'w', encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow (af)
        for i in range(0,len(p1)):
              writer.writerow ([p1[i]]+aa[i])
    


# In[9]:



#build classification model:
import numpy as np
import pandas as pd
train = pd.read_csv('nlppro.csv')


train


# In[15]:


columns=list(train)[1:]

ar1=set(columns)


# In[16]:


one=[]
for i in range(0,len(columns)):
        if(sum(train[columns[i]])<2):
            one.append(columns[i])

ar2=set(one)


# In[17]:


x = ar1
y = ar2

z = x.difference(y) 
z=list(z)
print(len(z))


# In[18]:


train=train[z]


# In[19]:


train.insert(0,'topic',p1)


# In[20]:


train


# In[21]:


train=train.iloc[np.random.permutation(len(train))]#shuffle


# In[22]:


#random forest:
from sklearn.model_selection import train_test_split
X = train.drop('topic',axis=1)
y = train['topic']


# In[24]:


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)

rfc_pred = rfc.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,rfc_pred))
print(classification_report(y_test,rfc_pred))


# In[55]:



def check(text):
    mj=stop(text)


    a=z
    b=[mj]
    aa=[]
    for i in range(0,len(b)):
            xx=[]
            for m in range(0,len(a)):
                xx.append(0)
            for j in range(0,len(b[i])):
                for k in range(0,len(a)):
                    if(b[i][j]==a[k]):
                        yy=a.index(a[k])
                        xx[yy]=1
            aa.append(xx)   

    #print(sum(aa[0]))
    fg=rfc.predict([aa[0]])[0]
    return fg


# In[59]:


check("ಪ್ರಪಂಚದಾದ್ಯಂತದ ವಿಜ್ಞಾನಿ ವೈರಸ್ ಮೂಲಗಳನ್ನು ಕಂಡುಹಿಡಿಯಲು ಪ್ರಯತ್ನಿಸುತ್ತಿದ್ದಾರೆ")

#scientist over the world are trying to figure the sources of virus
```

```python
from tkinter import *
class MyWindow:
    def __init__(self, win):
        self.lbl1=Label(win, text='Input Kannada text')
        self.lbl3=Label(win, text='Result')
        self.t1=Entry(bd=3)
        
        self.t3=Entry()
        self.btn1 = Button(win, text='Predict')
        
        self.lbl1.place(x=50, y=50)
        self.t1.place(x=200, y=50)
       
    
        self.b1=Button(win, text='Predict', command=self.add)
       
        self.b1.place(x=200, y=100)
        
        self.lbl3.place(x=50, y=200)
        self.t3.place(x=200, y=200)
    def add(self):
        self.t3.delete(0, 'end')
        num1=str(self.t1.get())
       
        result=check(num1)
        self.t3.insert(END, str(result))
    

window=Tk()
mywin=MyWindow(window)
window.title('Kannada News Category Classifier')
window.geometry("400x300+10+10")
window.mainloop()
```
